```python
# another method

[feature for feature in df.columns if df[feature].isnull().sum() > 0]
```

```python
# using seaborn heatmaps

sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# merge two datasets based on a common column

final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head(2)
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
# plot a pie chart if top 5 countries that uses Zomato

plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

The maximum number of transactions is from India then USA and the UK.

```python
final_df.columns
```

```python
# Exploration of Numerical features - 'Ratings'

final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Ratings Counts'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating", y="Ratings Counts", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Ratings Counts", data=ratings, hue='Rating color')
```

```python
color_palette = ['white', 'red', 'orange', 'yellow', 'green', 'green']
```

```python
sns.barplot(x="Aggregate rating", y="Ratings Counts", data=ratings, hue='Rating color', 
            palette=color_palette)
```

```python
sns.countplot(x="Rating color", data=ratings, palette=color_palette)
```
